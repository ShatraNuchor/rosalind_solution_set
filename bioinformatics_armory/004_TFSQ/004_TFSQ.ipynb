```python
# open and read file with open() func

file = open('tfsq_test.txt', 'r')
content = file.readlines()
file.close()

# looky into content

content
```

```python
# initiate an empty lists to keep identifiers and sequences

identifiers = list()
sequences = list()

# loop over content for identifiers

for i in range(0, len(content), 4):
    identifier = '>' + content[i].rstrip()[1:]
    identifiers.append(identifier)
    
# same for sequences

for i in range(1, len(content), 4):
    sequences.append(content[i].rstrip())
    
# print the lists

for i in range(len(identifiers)):
    print(identifiers[i])
    print(sequences[i])
```

```python
# we can improve this to one loop

identifiers = list()
sequences = list()

# loop over content for both identifiers and sequences

for i in range(0, len(content), 4):
    
    identifier = '>' + content[i].rstrip()[1:]
    identifiers.append(identifier)
    
    sequence = content[i + 1].rstrip()
    sequences.append(sequence)
    
# print the lists

for i in range(len(identifiers)):
    print(identifiers[i])
    print(sequences[i])
```

# 3. Implementation

```python
def tfsq(filename):
    
    '''
    input
        a fastq file
    process
        extracts only the sequence names and sequences in order
    output
        prints answer to console
        writes answer to a file
    '''
    
    # open output file
    output_file = open(f'{filename.split(".")[0]}_answer.txt', 'w')
    
    # print answer title
    print('\n\x1B[1mANSWER\x1B[0m\n______\n')
    
    # read input file
    with open(filename, 'r') as file:
        
        while True:
            identifier = file.readline()[1:]
            if len(identifier) == 0:
                break
                
            processed_identifier = '>' + identifier
            print(processed_identifier.rstrip())
            output_file.write(processed_identifier)
            
            sequence = file.readline()
            print(sequence.rstrip())
            output_file.write(sequence)
            
            file.readline()
            file.readline()
            
    # close output file
    output_file.close()
    print('\n\n#! The answer has been written into the file:',
          f'\x1B[1m./{filename.split(".")[0]}_answer.txt\x1B[0m\n')
```

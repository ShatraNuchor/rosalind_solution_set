# 2. Workspace

```python
# read the input file

with open('frmt_test.txt', 'r') as file:
    content = file.read()
    
# convert IDs to a list

ids = content.split()

# print

print(ids)
```

```python
# start NCBI session

from Bio import Entrez

Entrez.mail = '<your@mail.address>'
```

```python
# query the nucleotide db

handle = Entrez.efetch(db = 'nucleotide',
                       id = ids,
                       rettype = 'fasta')

records = handle.read()
```

```python
# look into the records

print(records)
```

```python
# look into raw records

records
```

```python
# this string can be splitted into identifiers and sequences

records = records.split('\n')
records
```

```python
# then the new list can be parsed

identifiers = list()
sequences = list()

sequence = ''

for item in records:
    if item.startswith('>'):
        if sequence != '':
            sequences.append(sequence)
        identifiers.append(item)
        sequence = ''
    elif item != '':
        sequence += item
sequences.append(sequence)
        
# print the first sequence and its identifier

print(identifiers[0])
print(sequences[0])
```

```python
# so we can find the shortest sequence's index

lenghts = list(map(lambda x: len(x), sequences))

idx = lenghts.index(min(lenghts))

print(idx)
```

```python
# use that index to identify shortest sequence and its name

print(identifiers[idx])
print(sequences[idx])
```

```python
# instead of all these steps, we can use biopython seqio fasta entry parser

from Bio import Entrez
from Bio import SeqIO

Entrez.mail = '<your@mail.address>'

handle = Entrez.efetch(db = 'nucleotide',
                       id = ids,
                       rettype = 'fasta')

SeqIO.parse(handle, 'fasta')
```

```python
# use that iterator

records = list(SeqIO.parse(handle, 'fasta'))

# look into records

records
```

```python
# we can look into the sequences or descriptions of each record

print(records[0].description)
print(records[0].seq)
```

```python
# we can measure the length of sequence

len(records[0].seq)
```

```python
# so we can loop the records to find the shortest one

shortest_seq = ''
shortest_desc = ''

ref_len = float('inf') # so any integer will be shorter than that for the first loop

for record in records:
    if len(record.seq) < ref_len:
        ref_len = len(record.seq)
        shortest_seq = record.seq
        shortest_desc = record.description
        
# print

print(ref_len)
print(shortest_desc)
print(shortest_seq)
```

```python
# but we can find the shortest sequence with one command also

shortest_record = min(records, key = lambda x: len(x.seq))
```

```python
# then extract necessary information

print(len(shortest_record.seq))
print(shortest_record.description)
print(shortest_record.seq)
```

```python
# but rosalind asks for a fasta output

import sys

SeqIO.write(shortest_record, sys.stdout, 'fasta')
```

# 3. Implementation

```python
def frmt(filename, mail_address):
    
    '''
    input
        a file contains genbank entry ids
        mail_address to connect NCBI
    process
        query genbank nucleotide db for ids to get fasta records
    output
        prints shortest sequence and its identifier to console
        writes shortest sequence and its identifier to a file
    '''
    
    from Bio import Entrez
    Entrez.mail = mail_address
    from Bio import SeqIO
    import sys
    
    # read input file and extract ids
    with open(filename, 'r') as file:
        content = file.read()
    ids = content.split()
        
    # query the db using ids
    handle = Entrez.efetch(db = 'nucleotide', id = ids, rettype = 'fasta')
        
    # parse the query results
    records = list(SeqIO.parse(handle, 'fasta'))
    
    # find the shortest sequence
    shortest_record = min(records, key = lambda x: len(x.seq))
    
    # result
    shortest_identifier = shortest_record.description
    shortest_seq = shortest_record.seq
    
    # print answer to console
    print('\n\x1B[1mANSWER\x1B[0m\n______\n')
    SeqIO.write(shortest_record, sys.stdout, 'fasta')
    
    # open file and write answer
    file = open(f'{filename.split(".")[0]}_answer.fasta', 'w')
    SeqIO.write(shortest_record, file, 'fasta')
    file.close()
    print('\n\n#! The answer has been written into the file:',
          f'\x1B[1m./{filename.split(".")[0]}_answer.fasta\x1B[0m\n')
```
